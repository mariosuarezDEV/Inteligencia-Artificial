# Predicción del salario según la experiencia

modelo para predecir el salario de una persona en función de sus años de experiencia laboral.

## Reto:

1. Entrena un modelo de regresión lineal con estos datos.


2. Evalúa el modelo con el puntaje

3. Predice el salario para 4, 6, 8, 10 y 12 años de experiencia.


4. Compara las predicciones con los valores reales esperados.

```python
import pandas as pd
datos = pd.read_csv("salario_train.csv")
```

```python
datos.info()
```

```python
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(x="Experiencia (años)", y="Salario (miles de $)", data=datos, hue="Salario (miles de $)", palette="coolwarm")
```

```python
from sklearn.linear_model import LinearRegression as LR
# Crear el modelo
modelo = LR()
```

```python
# Entrenar el modelo
X = datos["Experiencia (años)"]
y = datos["Salario (miles de $)"]

# Arreglo de 2D
X_procesada = X.values.reshape(-1, 1)
y_procesada = y.values.reshape(-1, 1)

modelo.fit(X_procesada, y_procesada)
```

```python
# Evaluar el modelo
modelo.score(X_procesada, y_procesada)
```

```python
test = pd.read_csv("salario_test.csv")
test.info()
```

```python
# Predecir
X_test = test["Experiencia (años)"]
X_test_procesada = X_test.values.reshape(-1, 1)
predicciones = modelo.predict(X_test_procesada) # predice los datos de y
```

```python
pd.DataFrame(predicciones).to_csv("predicciones.csv", header=None, index=None)

pd.DataFrame(predicciones).head()
```
