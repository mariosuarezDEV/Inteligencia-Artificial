# Predicción del precio de una casa en función del tamaño

Queremos predecir el precio de una casa basándonos en su tamaño en metros cuadrados.

$$
y = mX + b
$$

Donde:

$$
precio = m(tamaño) + b
$$

*m* es la pendiente (cuánto cambia el precio por metro cuadrado)

*b* es la intersección (precio base cuando el tamaño es 0).

## Crear los datos

```python
import pandas as pd
datos = pd.read_csv("datos_casas.csv")
```

## Visualizar los datos

```python
import seaborn as sb
# Recibe X (característica) e Y (etiqueta)
# La X es la entrada (con lo que se entrena el modelo)
# La Y es la salida (lo que se predice)
sb.scatterplot(x="Tamaño (m²)", y="Precio (miles de $)", data=datos) # Se grafica el tamaño vs el precio
```

### Otros parametros para ScatterPlot

*hue*: Categoriza los puntos por color segun una variable

    hue="Tipo de Propiedad (Columna)

*palette*: Define una paleta de colores cuando se usa **hue**

    palette="coolwarm"

```python
sb.scatterplot(x="Tamaño (m²)", y="Precio (miles de $)", data=datos, hue="Precio (miles de $)", palette="coolwarm") # Se grafica el tamaño vs el precio, con colores
```

## Entrenar el modelo

```python
from sklearn.linear_model import LinearRegression as LR

modelo = LR() # Se crea el modelo
```

```python
# Separar los datos en X y Y (entrada y salida) (característica y etiqueta)
X = datos["Tamaño (m²)"]
y = datos["Precio (miles de $)"]
```

```python
# Crear un arreglo de 2D para X
X_procesado = X.values.reshape(-1, 1)
# Crear un arreglo de 2D para y
y_procesado = y.values.reshape(-1, 1)
```

> Esto nos devuelve un arreglo de 2D en cada variable

```python
print(X_procesado)
```

```python
# Entrear el modelo con los datos de entrada y salida en 2D
modelo.fit(X_procesado, y_procesado)
```

```python
# Ver el score del modelo
print(modelo.score(X_procesado, y_procesado))
```

```python
# Probando el modelo haciendo predicciones
nuevos_tamanos = [[110], [135], [160], [185], [200]]
predicciones = modelo.predict(nuevos_tamanos)
```
